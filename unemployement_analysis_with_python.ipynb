```python
import pandas as pd

df = pd.read_csv("Unemployment in India.csv")
df.head()
```

```python
df.info()
df.isnull().sum()

# null values remove
df = df.dropna()

# column names check
df.columns
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df[' Date'], df[' Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate Trend")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
covid = df[df[' Date'].dt.year == 2020]

plt.figure(figsize=(10,5))
plt.plot(covid[' Date'], covid[' Estimated Unemployment Rate (%)'], color='red')
plt.title("Covid Impact on Unemployment (2020)")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Region', y=' Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.title("State-wise Unemployment Rate")
plt.show()
```

```python
Analysis of Unemployment Trends During the Covid-19 Pandemic

- The Covid-19 pandemic triggered a substantial surge in unemployment rates across the country.
- Strict lockdown measures severely disrupted economic activities, leading to widespread job losses in multiple sectors.
- In the post-2021 period, unemployment rates exhibited a gradual decline.
- This downward trend reflects a steady economic recovery and stabilization following the pandemic.
```
